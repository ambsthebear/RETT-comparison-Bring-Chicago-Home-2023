## Comparing RETT data across cities

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
RETT_info_df = pd.read_csv("RETT Numbers - Sheet1.csv")
```

```python
RETT_info_df.groupby(["Amount of Sale in Millions (USD)", "City"]).size().unstack().plot(kind="bar");
```

```python
RETT_info_df.pivot(index='Amount of Sale in Millions (USD)', columns='City', values='RETT').plot(kind="bar", figsize= (10,8), color= ["#c44601", "#f57600", "#8babf1", "#0073e6", "#e0ac00", "#5ba300", "#89ce00", "#e6308a"])
plt.title("Real Estate Transfer Tax Comparison", fontsize=24)
plt.xlabel("Amount of Sale in Millions (USD)", fontsize=18)
plt.ylabel("RETT percentage", fontsize=18)
plt.xticks(rotation="horizontal");
```
